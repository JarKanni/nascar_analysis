```python
import json
import pandas as pd
import numpy as np
```

```python
driver_id = 0
```

```python
# Get data into dataframe
with open('lap-times.json', 'r') as f:
    data = json.load(f)


df = pd.DataFrame(data['laps'][driver_id]['Laps'])

print(df)
```

```python
# get driver name
driver_name = pd.DataFrame(data['laps'][driver_id])
driver_name = driver_name['FullName'][driver_id]

driver_name = driver_name.lstrip('* ').rstrip(' #')

print(driver_name)
```

```python
# save as csv
driver_name = driver_name + '.csv'

print(driver_name)
df.to_csv(driver_name)
```

```python
for x in np.arange(43):

    driver_id = x
    
    # Get data into dataframe
    with open('lap-times.json', 'r') as f:
        data = json.load(f)
    
    
    df = pd.DataFrame(data['laps'][driver_id]['Laps'])
    
    print(df)
    
    # get driver name
    driver_name = pd.DataFrame(data['laps'][driver_id])
    driver_name = driver_name['FullName'][driver_id]
    driver_name = driver_name.lstrip('* ').rstrip(' #').rstrip('(i)')
    
    print(driver_name)
    
    
    # save as csv
    driver_name = driver_name + '.csv'
    
    print(driver_name)
    df.to_csv(driver_name)
```
